```python
import pandas as pd
df=pd.read_csv("poem_data6.csv")


df.head()

```

```python
from transformers import pipeline
from tqdm import tqdm

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_poem(poem_content):
    try:
        # Generate the summary using the pipeline
        input_length = len(poem_content.split())

        summary = summarizer(poem_content, max_length=input_length + 10, min_length=30, do_sample=False)[0]['summary_text']
        return summary
    except Exception as e:
        print(f"Error summarizing poem: {e}")
        return ""


# Apply the function to each row of the DataFrame
for index, row in tqdm(df.iterrows(), total=len(df), desc="Summarizing"):
    poem_content = row['Poem Content']
    summary = summarize_poem(poem_content)
    df.at[index, 'jist'] = summary
    df.to_csv("summarized_poems8.csv", index=False)  # Save DataFrame after each iteration


```
